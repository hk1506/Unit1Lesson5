## Background:

The newest product that young adults are raving about is the hydrocolloid acne patches. These are simple patches that contain hydrocolloid, a gel like compound that absorbs liquid. When the product is placed on a blemish or pimple for 6 hours, it absorbs the pus without leaving a scar. Due to its natural, safe, and easy solution to acne, it’s a popular item that people of all ages have been purchases. 
Hero Cosmetics launched in September, and their first product was a 36 count box of acne patches at the cost of $12.99 on Amazon. After a couple of months, they saw that some of their customers returned to purchase another box. Based on the number of repeat customers, they wondered if a second SKU should be created. This SKU would be the same product but with a greater number of patches in the box. 

## Problem: 
Will the launch of a new product (72 count acne patches) slow down the overall profits? The founders of Hero Cosmetics are afraid that if people purchase the 72 count, customers won’t purchase the patches as frequently as they do now. Hypothetically, if people purchase the 72 count, customers will take twice as long to finish a box. However, the founders also believe that if people have more patches for a cheaper unit price, then they will use the patches more willingly, without the fear of running out. 


## Potential Solution:
To get an insight to what may happen, an A/B experiment can be run. Two groups will be selected. These groups will be picked from the list of customers who have repeatedly purchased the product. One group will receive regular marketing emails about the 36 count. The second group will first receive a new product launch email and then regular marketing emails. But rather than marketing the 36 count, it will market both the 36 and 72 count boxes. 


## Before Rollout: 
Before starting the experiment, copies of the emails should be created. The emails should be as similar as possible to minimize distractions that might alter a consumer’s decision. The only method of marketing will be via email. 
The product design and box design will be exactly the same for products. A difference in product design or packaging design can change a user’s decision to purchase the 36 count over the 72 count (and vise versa). 
Each group should have 50 people and will be randomly selected into each group. A random generator will place people into either Group A or Group B. The 50 people who are picked will be picked from a list of people who have purchased the product more than once. In addition, the demographic will be of all ages and residents of cities all across the US. Over the next six month, both groups will receive between 6 to 12 emails. 


## Success Metric:
The success metric will based on the profits, which will be measure using t-tests and finding the p-values. 

## Secondary Success Metric:
A secondary metric will how soon after the email is sent will people purchase the product versus how long ago was their last purchase. Does a better deal draw people in to purchase an item? 


## Rollout Plan:
At the start of the experiment, both groups will receive a marketing email sent out the same day and time. Group A will get the standard email suggesting they purchase another box of acne patches. Group B will get a new product launch email. After two weeks, customers who have not purchased either products will receive a second email. Group A will a well designed marketing email listing the benefits of the hydrocolloid acne patches. And it will market the 36 count. Group B will receive the same email but it will provide an option of the 36 count or the 72 count. At the beginning of the next month, both groups will receive an email with positive customer reviews. Group A, of course, will get an email only advertising the 36 count. And Group B will get have the option of picking the 36 count or 72 count. For the next couple of months, similar emails will be sent out. Customers who purchase a product will receive a Thank You email. 


## Hypothesis:
Customers who purchase the 72 count will be less inclined to purchase the acne patches as frequently, and as a result, profits will decrease. 


## How to measure success:
At the end of the six months, we will look at the frequency in which customers purchased the 36 count versus the 72 count and calculate the profits. The primary measure of success will be test by looking at the t-test and p-values. If the p-value is < 0.05, then it will be an indicator that the null hypothesis was correct. 

The secondary metric will be measured by time. How long after the first email will customers purchase the product? Customers who purchase it within the first 48 hours are customers who purchased the product based on reminders. And customers who purchase product are customers who either need the product or have run out.
